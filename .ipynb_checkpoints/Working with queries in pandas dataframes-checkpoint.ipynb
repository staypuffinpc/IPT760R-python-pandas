# Query method

caveat: our argument will actually be a string
caveat: this method only works when col names don't have any spaces

```python
#import needed libraries
import pandas as pd
```

```python
bond = pd.read_csv("../data/pandas/jamesbond.csv",index_col="Film")
bond.sort_index(inplace = True)
bond.head(3) # show top 3 rows
```

```python
#need to change col headings so they don't have spaces
bond.columns = [column_name.replace(" ", "_") for column_name in bond.columns]
#check to see if the new headings are in place
bond.head(1)
```

```python
#query the data.  Supposedly the query method is faster for standard syntax, esp. on large datasets
bond.query("Actor=='Sean Connery'") #note that b/c the query is a string, I had to include a string inside of the string to search for a string.  Got it? ;)
```

```python
#query to find all movies with a specific director
bond.query("Director == 'Terence Young'")
```

```python
#query to find all movies that do NOT have David Niven
```

```python
bond.query("Actor != 'David Niven'")
```

```python
#use operations in a query
bond.query("Box_Office > 800")
```

```python
#query with multiple conditions.  Use 'and' and 'or'
bond.query("Actor == 'Sean Connery' or Bond_Actor_Salary >10")
```

```python
#pass a list in as the condition.  can use "not in" as well
bond.query("Actor in ['Timothy Dalton', 'George Lazenby']")
```
