```python
#import needed libraries
import pandas as pd
```

```python
bigmac = pd.read_csv("../data/pandas/bigmac.csv", parse_dates=['Date'])
```

```python
bigmac.info()
```

```python
#set multiple indexes
bigmac.set_index(keys = ['Date','Country'], inplace=True)
bigmac.head(3)
```

```python
#sort indexes
bigmac.sort_index(ascending=False, inplace=True)
bigmac.head(3)
```

```python
#notice that the index is now a multi-dimensional object
bigmac.index
```

```python
bigmac.index.names
```

```python
type(bigmac.index)
```

```python
bigmac.index[0]
```

```python
# use .sort_index() to increase efficiency
bigmac.sort_index(inplace=True)
```

## .get_level_values() method
Use to get a list of the values for specific columns or indexes

```python
bigmac.index.get_level_values(0)
```

```python
#can also use a textual value
bigmac.index.get_level_values('Date')
```

# .set_names() method
used to change the names of the index levels

```python
bigmac.index.set_names(["Day","Location"], inplace=True)
bigmac.head(3)
```

# .sort_index() method parameters

```python
#sort one index in one order and another index in another direction
bigmac.sort_index(ascending=[True, False], inplace=True)
bigmac.head(3)
```

# .loc() and .ix() methods to extract rows from a multi-index df

```python
# the problem is that .loc in pandas won't allow me to use a list.  So, I have to use a tuple.  e.g.,
bigmac.loc[("2010-01-01","Brazil")]
```

```python
#to extract a column, use a second argument (after the tuple)
bigmac.loc[("2010-01-01","Brazil"),"Price in US Dollars"]
```

```python
#another example
bigmac.loc[("2015-07-01", "Chile"), "Price in US Dollars"]
```

```python
#mix indexes and keys.  Be aware that .ix() is now deprecated
bigmac.ix[("2016-01-01", "China"), 0]
```

# .transpose() method
swap horizontal and vertical axes

```python
bigmac.transpose().head()
```

```python
#there is no 'inplace' attribute with .transpose().
bigmac = bigmac.transpose()
bigmac.head(3)
```

```python
# use .transpose() to return to prior format
bigmac = bigmac.transpose()
bigmac.head(3)
```

# .swaplevel() method
used to change the level positions in an index

```python
bigmac.swaplevel() #notice how Location and Day switched places
```

# .stack() method
moves the column-based index to a row-based index

```python
world = pd.read_csv("../data/pandas/worldstats.csv", index_col=["country","year"])
world.head(3)
```

```python
world.stack() #this is now a series, but with 3 indexes!
```

```python
#if you want to convert a series to a frame, use .to_frame()
stacked = world.stack().to_frame()
stacked.head(5)
```

```python
# .unstack() method will move a row index to cols
stacked.unstack()
```

```python
#run unstack again to move the next index into a column
stacked.unstack().unstack().unstack().head(5)
```

```python
#.unstack() parameters
stacked.unstack(2).head()
```

```python
stacked.unstack(1).head()
```

```python
#use negative arguments to reverse the direction of the index
stacked.unstack(-2).head() #note how this is the same as unstack(1) above.
```

```python
#you can also use keywords (name of indexes)
stacked.unstack("country").head()
```

```python
#provide a list for moving multiple columns
stacked.unstack(level=[1,0]).head() #moved these indexes to columns
```
