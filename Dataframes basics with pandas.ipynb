# Dealing with NaN values

```python
nba = pd.read_csv("../data/pandas/nba.csv")
nba.tail()
```

```python
# the .dropna() drops rows that have ANY null values at all.  BE CAREFUL in using this.
nba.dropna().tail()
```

```python
#change the "how" parameter to be more specific about what to remove
nba.dropna(how = "all").tail()
```

```python
# drop specific columns.  Change the axis to cols
nba.dropna(axis = 1)
nba
```

```python
#Get more specific.  Look for NaN in specific cols only
nba.dropna(subset = ['Salary', 'College']) #only drop if there is a null value in the Salary column
nba
```

# filling in NaN values with a pre-assigned value

```python
nba.fillna(0) #this replaces every NaN with a '0'
```

```python
#filter to only Salary and then fill it in
nba['Salary'].fillna(0, inplace = True)
```

```python
nba.head()
```

```python
#Fill college NaNs with "no college" string value
nba['College'].fillna('no college', inplace = True)
nba
```

# .astype() method
used to convert datatypes of a series from one type to another.
requires a series to not have any null values.  Therefore, you must convert those first.

```python
nba = pd.read_csv("../data/pandas/nba.csv")
```

```python
nba.dropna(how= "all", inplace=True)
nba.tail()
```

```python
#drop null values
nba.dropna( how = "all")
nba['Salary'].fillna(0, inplace = True)
nba['College'].fillna('no college', inplace = True)
```

```python
#check the datatypes
nba.dtypes
```

```python
#another way of checking data types.  Advantage of this method is that you can see how much memory is used.
nba.info()
```

```python
nba['Salary'] = nba['Salary'].astype('int')
```

```python
#convert age and number columns to ints
nba['Age'] = nba['Age'].astype('int')
nba.tail()
```

```python
#check for unique values and map all unique values to one location
nba['Position'].nunique()
```

```python
#turn the unique values from a specific column into a category type
nba['Position'] = nba['Position'].astype('category')
```

# Sorting dataframes

```python
nba = pd.read_csv("../data/pandas/nba.csv")
```

```python
#notice that chaining doesn't quite work here.  It resorts the resulting df each time
nba.sort_values('Team').sort_values('Salary').sort_values('Name') 
```

```python
#to sort by multiple columns, use a list.  It works for the ascending parameter, too
nba.sort_values(['Team','Salary','Name'], ascending=[True,False,True], inplace = True)
nba.tail(10)
```

```python
#sort_index() method works similarly on a df as it does on a series.  Use this to get the original shape back, since I changed it with the last command
nba.sort_index(inplace = True)
nba.tail(10)
```

# .rank() method to generate a new series

```python
#must first get rid of null values
nba.dropna(how="all")
nba['Salary'] = nba['Salary'].fillna(0).astype('int')
```

```python
#first need to extract the column of interest since this is a series operation
nba['Salary Rank'] = nba['Salary'].rank(ascending = False).astype('int')
nba.head(3)
```

```python
#sort by salary to make sure this is accurate
nba.sort_values(by = 'Salary',ascending=False)
```
