# .stack() method
moves the column-based index to a row-based index

```python
world = pd.read_csv("../data/pandas/worldstats.csv", index_col=["country","year"])
world.head(3)
```

```python
world.stack() #this is now a series, but with 3 indexes!
```

```python
#if you want to convert a series to a frame, use .to_frame()
stacked = world.stack().to_frame()
stacked.head(5)
```

```python
# .unstack() method will move a row index to cols
stacked.unstack()
```

```python
#run unstack again to move the next index into a column
stacked.unstack().unstack().unstack().head(5)
```

```python
#.unstack() parameters
stacked.unstack(2).head()
```

```python
stacked.unstack(1).head()
```

```python
#use negative arguments to reverse the direction of the index
stacked.unstack(-2).head() #note how this is the same as unstack(1) above.
```

```python
#you can also use keywords (name of indexes)
stacked.unstack("country").head()
```

```python
#provide a list for moving multiple columns
stacked.unstack(level=[1,0]).head() #moved these indexes to columns
```

```python
#populate NaNs that are a result from restructuring the shape of your df
stacked.unstack("year", fill_value=0)
```

# .pivot_table() method
ideal for aggregating a dataset and looking at it as a whole.  Behaves the same as a pivot table in MS Excel.

```python
#import foods data
foods = pd.read_csv("../data/pandas/foods.csv")
foods.head(3)
```

```python
#pivot it.  
## use the "values" parameter to determine how you want to pivot it.
## use "index" parameter to determine what will be in the rows
## aggfunc: determine what operation you want to run with this data
foods.pivot_table(values="Spend", index="Gender",aggfunc="mean")
```

```python
# check the sum of all that each gender has spent
foods.pivot_table(values="Spend", index="Gender",aggfunc="sum")
```

```python
#  check the sum of sales by item
foods.pivot_table(values="Spend", index="Item",aggfunc="sum")
```

```python
#To group first by gender, and then by item give two indexes
foods.pivot_table(values="Spend", index=["Gender","Item"],aggfunc="mean")
```

```python
#Let's look at sales by gender and item, but add a city column
foods.pivot_table(values="Spend", index=["Gender","Item"],columns = "City", aggfunc="mean")
```

```python
#for a multi-index column (e.g., frequency), use a list in the columns parameter
foods.pivot_table(values="Spend", index=["Gender","Item"],columns = ["Frequency","City"], aggfunc="mean")
```

### Other aggregation functions: 'count', 'max','min', 

```python
#find the max spent on each item per gender, per city
foods.pivot_table(values="Spend", index=["Gender","Item"],columns = "City", aggfunc="max")
```

```python
#find the least spent overall, per city
foods.pivot_table(values="Spend", index=["Item"],columns = "City", aggfunc="min")
```

```python
#count total # of orders of each item per city
foods.pivot_table(values="Spend", index=["Item"],columns = "City", aggfunc="count")
```

```python
#what was the median spent oin each item per city?
foods.pivot_table(values="Spend", index=["Item"],columns = "City", aggfunc="median")
```

# pd.melt() method
copnverts a dataframe to be more tabular

```python
quarterly = pd.read_csv("../data/pandas/quarters.csv")
quarterly
```

```python
#"melt" it
quarterly.melt()
```

notice how this moved everything to two columns.  We can get a bit more specific.

```python
# use id_vars to determine which value you want to maintain in place.  This can be used to help elongate your data
quarterly.melt(id_vars="Salesman")
```

```python
#same as above, but name the new columns
quarterly.melt(id_vars="Salesman",var_name="Quarter",value_name="Revenue")
```
