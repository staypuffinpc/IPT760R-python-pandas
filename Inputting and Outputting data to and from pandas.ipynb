```python
import pandas as pd
```

# pd.read_csv() to fetch a URL

```python
#NOTE: there are several places to find free/open data on the internet.  A great place to start is data.gov
url = "https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv"
baby_names = pd.read_csv(url)
baby_names.head(10)
```

```python
baby_names.info()
```

```python
#extract a series and convert it to a list
baby_names['Child\'s First Name'].tolist()
```

```python
baby_names['Child\'s First Name'].to_frame()
```

```python
#loop over series and combine them 
", ".join(str(name) for name in baby_names['Child\'s First Name'])
```

# Export CSV

```python
baby_names.to_csv("../data/pandas/NYC Baby Names.csv", index=False, columns=['Year of Birth', 'Gender'],
                 encoding="UTF-8") 
#index = False prevents the pandas default index from being exported
#columns allows you to choose which columns to export
# encoding allows you to set the language encoding.  This is important for languages with more diacritical marks
```
